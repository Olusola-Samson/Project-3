```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sb
%matplotlib inline
```

```python
#checking for empty columns
df.isnull().sum()
```

```python
#Reducing the datasets to contain the main features
newdf = pd.DataFrame()
newdf['Term'] = df['Term']
newdf['LoanStatus'] = df['LoanStatus'] 
newdf['BorrowerAPR %'] = df['BorrowerAPR'] * 100
newdf['BorrowerRate %'] = df['BorrowerRate'] * 100
newdf['LenderYield %'] = df['LenderYield'] * 100
newdf['EstimatedEffectiveYield %'] = df['EstimatedEffectiveYield'] * 100
newdf['EstimatedReturn %'] = df['EstimatedReturn'] * 100
newdf['EstimatedLoss %'] = df['EstimatedLoss'] * 100
newdf['DebtToIncomeRatio %'] = df['DebtToIncomeRatio'] * 100
newdf['CreditScore'] = ((df['CreditScoreRangeLower'] + df['CreditScoreRangeUpper']) / 2)
newdf['IncomeRange'] = df['IncomeRange'] 
newdf['StatedMonthlyIncome'] = df['StatedMonthlyIncome'] 
newdf['LoanOriginalAmount'] = df['LoanOriginalAmount'] 
newdf['LoanOriginationDate'] = df['LoanOriginationDate'] 
newdf['MonthlyLoanPayment'] = df['MonthlyLoanPayment'] 
newdf['LP_CustomerPayments'] = df['LP_CustomerPayments']
newdf['LP_NetPrincipalLoss'] = df['LP_NetPrincipalLoss']
newdf['IncomeVerifiable'] = df['IncomeVerifiable']

print(newdf.shape)
newdf.head(5)
```

```python
newdf.describe()
```

```python
newdf.info()
```

```python
#Filling the column with the mean
newdf['BorrowerAPR %'] = newdf['BorrowerAPR %'].fillna(newdf['BorrowerAPR %'].mean())

# dropping rows that have null values based on borrower apr
tf = newdf.dropna()
#reset index
tf = tf.reset_index(drop=True)
tf.info()
```

```python
var = {'LoanStatus' : ['Current','Completed','FinalPaymentInProgress','Past Due (1-15 days)','Past Due (16-30 days)',
'Past Due (31-60 days)','Past Due (61-90 days)','Past Due (91-120 days)','Past Due (>120 days)','Chargedoff','Defaulted' ]}

ordered_var = pd.api.types.CategoricalDtype(ordered = True, categories = var['LoanStatus'])
tf['LoanStatus'] = tf['LoanStatus'].astype(ordered_var)
tf['LoanStatus'].dtype
```

```python
tf.head()
```

```python
tf.describe()
```

```python
tf['CreditScore'].head(20)
```

```python
#Creating a new column to contain the order  of creditscore
CreditScore_Rating = []
a = 'Excellent'
b = 'Very good'
c = 'Good'
d = 'Fair'
e = 'Poor'
#iterating items in creditscore and giving them labels
for items in tf['CreditScore']:
    if items >= 800:
        CreditScore_Rating.append(a)
    elif int(items) in np.arange(740.0,799.99):
        CreditScore_Rating.append(b)
    elif int(items) in np.arange(670.0,739.9):
        CreditScore_Rating.append(c)
    elif int(items) in np.arange(580.0,669.9):
        CreditScore_Rating.append(d)
    else:
        CreditScore_Rating.append(e)
        
#print(CreditScore_Rating)
cr = pd.DataFrame({'CreditScore_Rating':CreditScore_Rating}) 
print(cr.head(10))
print(cr.isnull())
```

```python
#adding the new column to the dataset
tf = pd.concat([tf,cr], axis=1)
tf.head(20)
```

Next up are the factors that will determine how much lender's make and if they do make money

Simplified details about the factors beign explored:
 > BorrowerAPR: The Borrower's Annual Percentage Rate (APR) for the loan.
 
 > BorrowerRate: The Borrower's interest rate for this loan in percentage.
 
 > LenderYield: The Lender yield on the loan. Lender yield is equal to the interest rate on the loan less the servicing fee.
 
 > EstimatedEffectiveYield: Effective yield is equal to the borrower interest rate (i) minus the servicing fee rate, (ii) minus estimated uncollected interest on charge-offs, (iii) plus estimated collected late fees. Applicable for loans originated after July 2009 in percentage.


```python
#plots for the different intrest rates
fig, ((ax,ax1),(ax2,ax3)) = plt.subplots(2,2,figsize=[14,8], constrained_layout = True)

binsize = 10
def bins(col):
    bins = np.arange(0, col.max()+binsize, binsize)
    return bins

axbin = bins(tf['BorrowerAPR %'])
ax.hist(data = tf, color ='r', x = 'BorrowerAPR %', bins = axbin)
ax.set(title = "BorrowerAPR in percentage",  xlabel ="BorrowerAPR %")

ax1bin = bins(tf['BorrowerRate %'])
ax1.hist(data = tf, color ='g', x = 'BorrowerRate %', bins = ax1bin)
ax1.set(title = "BorrowerRate in percentage",  xlabel ="BorrowerRate %")

ax2bin = bins(tf['LenderYield %'])
ax2.hist(data = tf, color ='y', x = 'LenderYield %', bins = ax2bin)
ax2.set(title = "LenderYield in percentage", xlabel ="LenderYield %")

ax3bin = bins(tf['EstimatedEffectiveYield %'])
ax3.hist(data = tf, color ='purple', x = 'EstimatedEffectiveYield %', bins = ax3bin)
ax3.set(title = "EstimatedEffectiveYield in percentage",  xlabel ="EstimatedEffectiveYield %")

plt.suptitle("Histogram  of the Intrest Rates in percentages".title(), fontsize=17, weight = 'bold')


plt.show();
```

**As illustrated above a higher section of people's BorrowerAPR falls between 10 to 40% and a lower section go below that .
In the case of the BorrowerRate a higher percent falls between the 10 to 30% range while the rest are below or higher than that.
While for LenderYield a fairly decent amount fall between 0 to 10% while most fall between 10 to 30% and the rest belong to a smaller demographic
Finally, for EstimatedEffectiveYield like previous analysis most fall between the 10 to 30% range and few fall between 0 to 10% while little or nothing fall between 30 to 40%**
